```python
# library for data manipulation
import pandas as pd
import numpy as np

# library for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

<h3>Checking and Cleaning the data</h3>

```python
# reading the csv file into a dataframe
diamonds_df = pd.read_csv("diamonds.csv")
```

```python
# taking the names of the columns
diamonds_columns = list(diamonds_df.columns)
```

```python
# giving the index column a name
diamonds_df.columns = ['index'] + diamonds_columns[1:]
```

```python
# assigning the new column names to the dataframe
diamonds_df.columns
```

```python
# setting the index
diamonds_df.set_index("index", inplace = True)
```

```python
# checking the head of the dataframe
diamonds_df.head()
```

```python
# around 25 percentile of the diamonds cost less than 650 dollars
# around 50 percentile of the diamonds cost less than 2500 dollars
# around 75 percentile of the diamonds cost less than 5700 dollars
# most of the diamonds cost less than 12500
#
# diamonds whose price are in out liers
len(diamonds_df[diamonds_df['price'] > 11875])
```

```python
# effect of carat and cut on the price of the diamonds
sns.lmplot(data=diamonds_df, x='carat', y='price', hue='cut',
              hue_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], scatter_kws={'alpha':0.1, 's':5, 'marker':'D'},
              line_kws={'lw':1})

plt.show()
```

```python
# worst to best -----> Fair, Good, Very Good, Premium, Ideal
#     from the plot we can see that the price increase with carat is the lowest for 'Fair' cuts followed by 'Good' cuts.
#     for 'Very Good' and 'Premium' cuts there seem to be not much difference.
#     'Ideal' cuts seem to have a slight higher increase in price with carat than the 'very good' and 'premium'
# cuts.
#     In conclusion, apart from 'Fair' cutthe effect of the cut of a diamond doesn't seem to have a significant effect on the price
# of the diamond  as we can see a lot of overlap in the plot.
```

```python
# effect of carat and clarity on price
sns.lmplot(data=diamonds_df, x='carat', y='price', hue='clarity',
              hue_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'], scatter_kws={'alpha':0.1, 's':5, 'marker':'D'},
              line_kws={'lw':1})

plt.show()
```

```python
# worst to best -----> I1, SI2, SI1, VS2, VS1, VVS2, VVS1, IF.
#     Price increase per carat seem to be the lowest for 'I1'.
#     This is followed by 'SL2' and 'SL1' where 'SL1' is slightly ahead of 'SL2'.
#     This is followed by 'VS2' and 'VS1' where 'VS1' is slightly ahead of 'VS2'.
#     This is followed by 'VVS2' and 'VvS1' where 'VVS1' is slightly ahead of 'VVS2'.
#     This is followed by 'IF'
#     In Conclusion, the clarity of the diamond has a significant impact on the price of a diamond.
# But the difference in the increase in the price between 'VS2' and 'VS1' seem to be less, same in the
# case between 'VVS2' and 'VVS1'.
```

```python
# effect of carat and color on the price of the diamonds
#sns.stripplot(data=diamonds_df, x='carat', y='price', hue='color',alpha=0.7, marker='D',
#              hue_order=['J', 'I', 'H', 'G', 'F', 'E', 'D'])
sns.lmplot(data=diamonds_df, x='carat', y='price', hue='color',
              hue_order=['J', 'I', 'H', 'G', 'F', 'E', 'D'], scatter_kws={'alpha':0.1, 's':5, 'marker':'D'},
              line_kws={'lw':1})
plt.show()
```

```python
# Worst to Best -----> J, I, H, G, F, E, D
#   The color that seem to have the least increase in the price with carat is "J".
#   Next come the colors "I" and "H" which have a similar increase in price with carat.
#   This is followed by "G", "F", "E" and "D" where all these have similar increase in price.
#   In conclusion, the color of the diamond doesn't seem to have a significatn role in determining the price of the diamond.
```
